# Check if someone has taken RCR at CU Boulder

```python
def checkonepersonfortraining(FirstN, LastN, dftocheck):
    dftocheck_person = dftocheck[dftocheck["person"].str.contains(LastN, na=False, case=False)]
    dftocheck_person = dftocheck_person[dftocheck_person["person"].str.contains(FirstN, na=False, case=False)]
    hasalternativenames = len([col for col in dftocheck_person.columns if col.find("alternative_name")>-1])
    if hasalternativenames==1:
        dftocheck_person2 = dftocheck[dftocheck["alternative_name"].str.contains(LastN, na=False, case=False)]
        dftocheck_person2 = dftocheck_person2[dftocheck_person2["alternative_name"].str.contains(FirstN, na=False, case=False)]
        return max(dftocheck_person.shape[0], dftocheck_person2.shape[0])
    else:
        return dftocheck_person.shape[0]

def checkneedtraining(row):
    grant_still_active = row["grant_still_active"]
    faculty = row["faculty"]
    training_inlast4years = row["training_inlast_4years"]
    Program_isSHORT = row["Program_isSHORT"]
    Program_isSummer = row["Program_isSummer"]
    #"Short Courses"
    timetotrain= False
    if grant_still_active:
        if faculty==False:
            if training_inlast4years==False:
                if Program_isSHORT==False:
                    if Program_isSummer==False:
                        timetotrain= True
    return timetotrain
```
